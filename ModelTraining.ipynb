```python
# Above listed are object columns in both datasets
# getting unique values for object columns

obj_cols = list(df_train.dtypes[df_train.dtypes == 'object'].index)
obj_cols
```

```python
for x in obj_cols:
    print('Unique values in df_train', [x], 'column :','\n', df_train[x].unique(),'\n')
```

```python
# above mentioned are column and unique values
# label encoding object columns in Train and Test both datasets

from sklearn.preprocessing import LabelEncoder
l_encode = LabelEncoder()
for x in obj_cols:
    df_train[x] = l_encode.fit_transform(df_train[x])
    df_test[x] = l_encode.fit_transform(df_test[x])
    
```

```python
# Checking unique values after label encoding
for x in obj_cols:
    print('Unique values in df_train', [x], 'column :','\n', df_train[x].unique(),'\n')
```

```python
# Scaling the df_train
# First splitting target(y) from df_train

y = df_train['y']
x = df_train.drop(['y'],axis = 1)
print(y.head())
print(x.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
```

```python
# scaling x data
x = scale.fit_transform(x)
```

```python
# visualising for confirmation
pd.DataFrame(x)
```

```python
pd.DataFrame(x).describe()
```

```python
# Dimensionality reduction using PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 0.95)  # Dimensionality reduction to keep 99% variance 
pca.fit(x)                      # Fitting train dataset (x) into PCA 
pca.explained_variance_ratio_   # Getting columns needed to keep 99% variance
```

```python
pca.explained_variance_.shape
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
# 76 columns carry 95% variance and hence dimensionality can be reduced from '365' to '76'
# Getting transformed train (x) dataset 

x_trans = pca.transform(x)
x_trans.shape
```

```python
# Performing train test split  
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_trans, y, test_size = 0.25, random_state = 5)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Model building using XGBoost
from xgboost import XGBRegressor
xgb_model = XGBRegressor()
```

```python
xgb_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
train_pred = xgb_model.predict (x_train)
test_pred = xgb_model.predict(x_test)
print('R2 score for x_train = ', r2_score (y_train, train_pred))
print('R2 score for x_test =', r2_score(y_test, test_pred))
```
